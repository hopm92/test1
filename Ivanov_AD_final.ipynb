```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
import time

# Создаем графический интерфейс с помощью Tkinter
root = tk.Tk()
root.title('Сортировка чисел')
root.geometry('400x300')

# Создаем поле ввода текста
input_label = ttk.Label(root, text='Введите последовательность чисел через запятую:')
input_label.pack()
input_entry = ttk.Entry(root)
input_entry.pack()

# Создаем раскрывающийся список для выбора сортировки
sort_label = ttk.Label(root, text='Выберите тип сортировки:')
sort_label.pack()
sort_combobox = ttk.Combobox(root, values=['По возрастанию', 'По убыванию'])
sort_combobox.pack()

# Создаем поле вывода надписи и кнопку
output_label = ttk.Label(root, text='Отсортированная последовательность:')
output_label.pack()
output_text = tk.Text(root, height=5)
output_text.pack()
time_label = ttk.Label(root, text='Затраченное время:')
time_label.pack()
time_text = ttk.Label(root, text='')
time_text.pack()

def sort_numbers():
    # Получаем введенную последовательность чисел
    input_str = input_entry.get()
    try:
        # Преобразуем строку в список чисел
        numbers = [int(num.strip()) for num in input_str.split(',')]
        # Получаем выбранный тип сортировки
        sort_type = sort_combobox.get()
        # Измеряем затраченное время
        start_time = time.time()
        for i in range(0, 1000000):
            pass
        # Сортируем список чисел
        if sort_type == 'По возрастанию':
            numbers.sort()
        else:
            numbers.sort(reverse=True)
        end_time = time.time()
        # Выводим отсортированный список чисел и время затраченное на сортировку
        output_text.delete('1.0', tk.END)
        output_text.insert(tk.END, ', '.join(map(str, numbers)))
        time_text.config(text='{} секунд'.format(round(end_time - start_time, 10)))
    except ValueError:
        # Если введены некорректные данные, выводим сообщение об ошибке
        messagebox.showerror('Ошибка', 'Введите корректную последовательность чисел через запятую')

# Создаем кнопку для запуска сортировки
button = ttk.Button(root, text='Start', command=sort_numbers)
button.pack()

root.mainloop()
```
